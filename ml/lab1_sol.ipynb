```python
import os
import numpy as np
import skimage
import matplotlib.pyplot as plt
```

```python
# Metoda 1
images1 = []
for filename in os.listdir('images'):
    images1.append(np.load(os.path.join('images',filename)))
    
images1 = np.array(images1) # (9, 400, 600)
print(images1.shape)

# Metoda 2
images2 = np.zeros(shape=(9,400,600))
for i,filename in enumerate(os.listdir('images')):
    images2[i] = np.load(os.path.join('images',filename))
print(images2.shape)

print(f'Sunt egale? {(images1 == images2).all()}')
```

```python
# Metoda 1
print(f'index max = {np.argmax(images1.sum(axis=(1,2)))}')

# Metoda2
max_img = 0
max_idx = -1

for idx,image in enumerate(images1):
    if image.sum() > max_img:
        max_img = image.sum()
        max_idx = idx
        
print(f'index max = {max_idx}')
```
